```python
# Add your data file to plot
data_file = "grok_asg_anomaly.txt"
data_file1 = "expect_grok_asg_anomaly.txt"
data_file2 = "predict_grok_asg_anomaly.txt"
data_file3 = "dtw_similarity_grok_asg_anomaly.txt"
dataframe = pandas.read_csv(data_file, header=None, names=["value"])
dataframe1 = pandas.read_csv(data_file1, header=None, names=["value"]) 
dataframe2 = pandas.read_csv(data_file2, header=None, names=["value"])
dataframe3 = pandas.read_csv(data_file3, header=None, names=["value"])

def make_offset(y,offset):
    new_y = np.zeros(len(y)+offset)
    for i in range(0,len(y)):
        new_y[i+offset] = y[i]
    return new_y

# x = np.array(dataframe['timestamp'])
values = np.array(dataframe['value'])
values1 = np.array(dataframe1['value'])
values2 = np.array(dataframe2['value'])
values3 = np.array(dataframe3['value'])
size = len(values)
size1 = len(values1)
size2 = len(values2)
size3 = len(values2)
x = np.arange(1,size+1)
y = np.array(values)

x1 = np.arange(1,size1+1)
y1 = np.array(values1)

x1 = x
y1 = make_offset(y1,int(360))

x2 = np.arange(1,size2+1)
y2 = np.array(values2)

x2 = x
y2 = make_offset(y2,int(360))

x3 = np.arange(1,size3+1)
y3 = np.array(values3)

x3 = x
y3 = make_offset(y3,int(360))

trace = {"x": x,
         "y": y,
         "mode": 'lines',
         "name": 'Value'}
trace1 = {"x": x1,
         "y": y1,
         "mode": 'lines',
         "name": 'expexted'}
trace2 = {"x": x2,
         "y": y2,
         "mode": 'lines',
         "name": 'predicted'}
trace3 = {"x": x3,
         "y": y3,
         "mode": 'lines',
         "name": 'dtw'}
traces = [trace,trace1,trace2,trace3]
layout = dict(title = "Data plot : "+data_file,
              xaxis = dict(title = 'X'),
              yaxis = dict(title = 'Value')
             )
fig = dict(data=traces, layout=layout)
iplot(fig)
print("size values: "+str(size))
print("size expected : "+str(size1))
print("size predicted : "+str(size2))
print("size dtw : "+str(size3))
```
